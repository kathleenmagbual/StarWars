```python
col_names = ["Very unfavorably", "Somewhat unfavorably","Neutral", "Somewhat favorably", "Very favorably"] 
results = {"R2D2": R2D2_final.iloc[:,1], 'C3P0': C3P0_final.iloc[:,1] } 
def survey(results, col_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('PiYG')(np.linspace(0.1, 0.90, data.shape[1]))
    fig, ax = plt.subplots(figsize=(9.9, 3.7))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())
    for i, (colname, color) in enumerate(zip(col_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.9, label=colname, color=color)
        xcenters = starts + widths / 2
        r, g, b, _ = color
        text_color = 'black' 
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(col_names), bbox_to_anchor=(0, -0.15),
              loc='lower left', fontsize='small')
    return fig, ax
     
survey(results, col_names) 
plt.title("Favorablity Ratings between C-3P0 and R2D2 (in %)")

fig.tight_layout()
plt.show()
fig.savefig('plot5.pdf')
```

# Chi Square Correlation

### Han Solo and Gender

```python
Han = starwars_clean.iloc[:, [15,33]]
Han_crosstab = pd.crosstab(Han['Han_Solo'], 
                            Han['Gender'],  
                               margins = False)
Han_crosstab
```

```python
Han_crosstab = Han_crosstab.iloc[[0,2,3,4,5,6,], [0,1,2]]
from scipy.stats import chi2_contingency
# we create contingency table same as above
table = [[Han_crosstab.iloc[:,0]], [Han_crosstab.iloc[:,1]]]
# Get chi-square value , p-value, degrees of freedom, expected frequencies using the function chi2_contingency
stat, p_hanG, dof, expected = chi2_contingency(table)
# select significance value
alpha = 0.05
# Determine whether to reject or keep your null hypothesis
print('significance=%.3f, p=%.3f' % (alpha, p_hanG))
if p_hanG <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
```

```python
Sw_Chars = ["Han_Solo", "Luke_Skywalker", "PrincessLeiaOrgana",
              "Anakin_Skywalker", 'Obi_WanKenobi',
                  'Emperor_Palpatine', 'Darth_Vader']
              
              #'Lando_Calrissian', 'Boba_Fett', 'C-3P0', 'R2D2',
               #   'JarJar_Binks', 'Padme Amidala', 'Yoda'
Demogs = ["Gender", "Age", "Income",
           "Education", "Location"]

corr = np.array([[p_hanG.round(3), p_HanA.round(2) , p_HanI.round(2), p_HanE.round(2), p_HanL.round(2)],
                    [p_LukeG.round(3), p_LukeA.round(2), p_LukeI.round(2), p_LukeE.round(2), p_LukeL.round(2)],
                    [p_LeiG.round(3), p_LeiA.round(2), p_LeiI.round(2), p_LeiE.round(2), p_LeiaL.round(2)],
                    [p_AnaG.round(3), p_AnaA.round(2), p_AnaI.round(2), p_AnaE.round(2), p_AnaL.round(2)],
                    [p_ObiG.round(3), p_ObiA.round(2), p_ObiI.round(2), p_ObiE.round(2), p_ObiL.round(2)],
                    [p_EmpG.round(3), p_EmpA.round(2), p_EmpI.round(2), p_EmpE.round(2), p_EmpL.round(2)],
                    [p_DVG.round(3), p_DVA.round(2), p_DVI.round(2), p_HanI.round(2), p_DVL.round(2)]])


Sw_Chars2 = ['Lando_Calrissian', 'Boba_Fett', 'C-3P0', 'R2D2',
            'JarJar_Binks', 'Padme Amidala', 'Yoda']
Demogs2 = ["Gender", "Age", "Income",
           "Education", "Location"]

corr2 = np.array([[p_LCG.round(2), p_LCA.round(2) , p_LCI.round(2), p_LCE.round(2), p_LCL.round(2)],
                    [p_BobaG.round(2), p_BobA.round(2), p_BobaI.round(2), p_BobaE.round(2), p_BobaL.round(2)],
                    [p_C3G.round(2), p_C3A.round(2), p_CPI.round(2), p_C3E.round(2), p_C3L.round(2)],
                    [p_R2G.round(2), p_R2A.round(2), p_R2I.round(2), p_R2E.round(2), p_R2L.round(2)],
                    [p_JJG.round(2), p_JJA.round(2), p_JJI.round(2), p_JJE.round(2), p_JJL.round(2)],
                    [p_PadG.round(2), p_PadA.round(2), p_PadmeI.round(2), p_PadmeE.round(2), p_PadL.round(2)],
                    [p_YodaG.round(2), p_YodaA.round(2), p_YodaI.round(2), p_YodaI.round(2), p_YodaL.round(2)]])


fig, axes = plt.subplots(ncols=2, figsize=(14, 6))
ax, ax2 = axes

def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")

    # We want to show all ticks...
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    # ... and label them with the respective list entries.
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)
    

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right",
             rotation_mode="anchor")

    # Turn spines off and create white grid.
    for edge, spine in ax.spines.items():
        spine.set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)
    

    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.3f}",
                     textcolors=["black", "white"],
                     threshold=None, **textkw):

    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            kw.update(color=textcolors[int(im.norm(data[i, j]) > threshold)])
            text = im.axes.text(j, i, valfmt(data[i, j], None), **kw)
            texts.append(text)

    return texts

import matplotlib

im, cbar = heatmap(corr, Sw_Chars, Demogs, ax=ax,
                   cmap='Wistia', cbarlabel="Chi-Square p-value between 2 variables")
texts = annotate_heatmap(im, valfmt="{x:.2f}")

#Plot2
im, cbar = heatmap(corr2, Sw_Chars2, Demogs2, ax=ax2,
                   cmap= 'Wistia', cbarlabel="Chi-Square p-value between 2 variables")
texts = annotate_heatmap(im, valfmt="{x:.2f}")

fig.tight_layout()
fig = plt.gcf()
plt.show()
plt.draw()
fig.savefig('plot6.pdf', dpi =100)
```
